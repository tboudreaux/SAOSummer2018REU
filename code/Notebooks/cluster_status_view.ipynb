```python
# PATH = "../../simulations/runTests/ExternalPotentialTest2/N1k_R3_K_xtern/cluser_status.csv"
# PATH = "../intraRunOutput/ExternalPotentialTest3/N1k_R3_K_xtern_V/cluser_status.csv"
# PATH = "../intraRunOutput/ExternalPotentialTest4/N1k_R3_King_xtern_V3/cluser_status.csv"
# PATH = "../intraRunOutput/ExternalPotentialTest5/N1K_R3_K_MWP/cluser_status.csv"
# PATH = "../../simulations/runTests/LongTermEvolutionTest/N10K_R3_K_t3/cluser_status.csv"
# PATH = "../../simulations/runTests/OrbitTest/N1k_R3_K_t3_V30/cluser_status.csv"
# PATH = "../../simulations/runTests/OrbitTest/N1k_R3_K_t2_V50/cluser_status.csv"
# PATH = "../../simulations/runTests/OrbitTest/N1k_R3_K_T2/cluser_status.csv"
# PATH = "../../simulations/runTests/OrbitTest/N1k_R3_K_xtern/cluser_status.csv"
# PATH = "../../simulations/runTests/OrbitTest/N1k_R3_K_t3_V0/cluser_status.csv"
PATH = "../../simulations/externalSimulations/TestSimulations/preCulsterTests/OLD_N3k_K_T10000_t327-06-2018_07:54:46/cluster_status.csv"

base_filename = "path_{}".format(PATH.split('/')[-2])
```

```python
save = False
```

```python
key = 'RDENS'
df = pd.read_csv(PATH, sep=r'\s+', engine='python')
print("Evolved for {:0.2f}Myr".format(df.iloc[-1]['TTOT']*df.iloc[-1]['TSTAR']))
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(df["{}1".format(key)], df["{}2".format(key)], df["{}3".format(key)], 'k-', linewidth=1)
ax.plot([df.iloc[0]["{}1".format(key)]], [df.iloc[0]["{}2".format(key)]], [df.iloc[0]["{}3".format(key)]], 'g+', markersize=10)
ax.plot([df.iloc[-1]["{}1".format(key)]], [df.iloc[-1]["{}2".format(key)]], [df.iloc[-1]["{}3".format(key)]], 'r+', markersize=10)

ax.set_xlabel('X [ly]', fontsize=20, labelpad=15)
ax.set_ylabel('Y [ly]', fontsize=20, labelpad=15)
ax.set_zlabel('Z [ly]', fontsize=20)

ax.tick_params(axis='both', which='major', labelsize=15, length=12, direction='in')
ax.tick_params(axis='both', which='minor', length=6, direction='in')

ax.view_init(45, 300)

if save:
    filename = "{}_{}.pdf".format(base_filename, 'XYZ')
    plt.savefig('Figures/PaperFigures/{}'.format(filename), bbox_inches='tight')
```
