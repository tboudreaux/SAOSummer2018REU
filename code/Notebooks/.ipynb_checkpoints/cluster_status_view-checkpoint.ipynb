```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import MultipleLocator
from matplotlib.ticker import AutoMinorLocator


plt.rc('text', usetex=True)
plt.rc('font', family='serif')

%matplotlib inline
```

```python
PATH = "../intraRunOutput/ExternalPotentialTest2/N1k_R3_K_xtern/cluser_status.csv"
# PATH = "../intraRunOutput/ExternalPotentialTest3/N1k_R3_K_xtern_V/cluser_status.csv"
# PATH = "../intraRunOutput/ExternalPotentialTest4/N1k_R3_King_xtern_V3/cluser_status.csv"
# PATH = "../intraRunOutput/ExternalPotentialTest5/N1K_R3_K_MWP/cluser_status.csv"
```

```python
key = 'RDENS'
df = pd.read_csv(PATH, sep=r'\s+', engine='python')
print("Evolved for {:0.2f}Myr".format(df.iloc[-1]['TTOT']*df.iloc[-1]['TSTAR']))
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(df["{}1".format(key)], df["{}2".format(key)], df["{}3".format(key)], 'o-')
ax.plot([df.iloc[0]["{}1".format(key)]], [df.iloc[0]["{}2".format(key)]], [df.iloc[0]["{}3".format(key)]], 'go')
ax.plot([df.iloc[-1]["{}1".format(key)]], [df.iloc[-1]["{}2".format(key)]], [df.iloc[-1]["{}3".format(key)]], 'ro')

ax.tick_params(axis='both', which='major', labelsize=15, length=12, direction='in')
ax.tick_params(axis='both', which='minor', length=6, direction='in')

ax.view_init(30, 45)
```

```python
base_filename = "path_{}".format(PATH.split('/')[-2])
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.plot(df.RDENS1, df.RDENS2, 'k-', linewidth=1)

ax.set_xlabel('X [ly]', fontsize=20)
ax.set_ylabel('Y [ly]', fontsize=20)

XMinor = MultipleLocator(125)
YMinor = MultipleLocator(125)

XMinor = AutoMinorLocator()
YMinor = AutoMinorLocator()

ax.xaxis.set_minor_locator(XMinor)
ax.yaxis.set_minor_locator(YMinor)

ax.tick_params(axis='both', which='major', labelsize=20, length=12, direction='in', top=True, right=True)
ax.tick_params(axis='both', which='minor', length=6, direction='in', top=True, right=True)

filename = "{}_{}.pdf".format(base_filename, 'XY')
plt.savefig('Figures/PaperFigures/{}'.format(filename), bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.plot(df.RDENS2[df['TTOT'] < 50], df.RDENS3[df['TTOT'] < 50], 'k', linewidth=1)

ax.set_xlabel('Y [ly]', fontsize=20)
ax.set_ylabel('Z [ly]', fontsize=20)

XMinor = AutoMinorLocator()
YMinor = AutoMinorLocator()

ax.xaxis.set_minor_locator(XMinor)
ax.yaxis.set_minor_locator(YMinor)

ax.tick_params(axis='both', which='major', labelsize=20, length=12, direction='in', top=True, right=True)
ax.tick_params(axis='both', which='minor', length=6, direction='in', top=True, right=True)

filename = "{}_{}.pdf".format(base_filename, 'YZ')
plt.savefig('Figures/PaperFigures/{}'.format(filename), bbox_inches='tight')
```
